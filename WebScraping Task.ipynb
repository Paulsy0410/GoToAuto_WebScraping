```python

from selenium import webdriver
pageNo=2
carDetails={}
carName=[]
price=[]

for currentPage in range(1,pageNo+1):  #loop for pages
    
    driver = webdriver.Chrome(executable_path=r"/home/user/Downloads/chromedriver_linux64/chromedriver")
    driver.get('https://www.gotoauto.ca/inventory/?pg='+str(currentPage))
    htmlPagNo='page'+str(currentPage)+'.html' #variable for pageNumber rotation
    with open(htmlPagNo, 'w+') as f: #Writting html page
        f.write(driver.page_source)
        f.close()
        driver.close()
        
        with open(htmlPagNo) as fp:
            soup = BeautifulSoup(fp)
            
        carlist1 = soup.find_all(class_="list")
        
        for i in carlist1: # loop for cars
            #print(i)
            print('\n')
    
            print(i.find('h4', attrs={'class':'name desc_l5'}).text) #Scraping car details
            carName1=i.find("h4",attrs={'class':'name desc_l5'}).text
            carNameUpd= carName1.replace('\t'," ").replace('\n'," ") #replacing "\t" &"\n"
            newNameVal= " ".join(carNameUpd.split())
            carName.append(newNameVal) #Appending car name to list
            
    
            print(i.find('span', attrs={'class':'price'}).text) #Scraping price details
            price.append(i.find('span', attrs={'class':'price'}).text) #appending price to list
                         
                     
```

# TASK 1

```python
carDetails['Car_name'] =carName
carDetails['MSRP'] =price


import pandas as pd #DataFrame creation 
data= pd.DataFrame(carDetails)
```

```python
from selenium import webdriver
pageNo=1


for currentPage in range(1,pageNo+1):  #loop for pages
    driver = webdriver.Chrome(executable_path=r"/home/user/Downloads/chromedriver_linux64/chromedriver")
    driver.get("https://www.gotoauto.ca/inventory/")
    with open('page.html', 'w+') as f:
        f.write(driver.page_source)
        f.close()
        driver.close()
        with open("page.html") as fp:
            soup = BeautifulSoup(fp)
            
            carlist1 = soup.find_all(class_="list")
            for i in carlist1: # loop for cars
                #print(i)
                print('\n')
            
                print(i.find('h4', attrs={'class':'name desc_l5'}).text) #Scraping car details
                carName1=i.find("h4",attrs={'class':'name desc_l5'}).text
                carNameUpd= carName1.replace('\t'," ").replace('\n'," ") #replacing "\t" &"\n"
                newNameVal= " ".join(carNameUpd.split())
                carName.append(newNameVal) #Appending car name to list
            
            
                print("https://www.gotoauto.ca"+ i.find("a")['href']) #scraping the Hyperlinks
                headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
                subDetails = requests.get("https://www.gotoauto.ca"+ i.find("a")['href'],headers=headers)
                subsoup = BeautifulSoup(subDetails.content,'html.parser')
                para = subsoup.find(class_='col col-12 note_desc_col finance_active').text #scraping the description
                print(para)
                textFileName= str(newNameVal +".txt") # converting to text file 
                with open(textFileName, 'w+') as f:
                    f.write(para)
                    f.close()
                    
            
            
            
    
```
